##12. Copy vs View

A view shares the original data, while a copy gets its own brand-new data.



```python
# Copy
arr = np.array([1, 2, 3])

copy_arr = arr.copy()

copy_arr[0] = 100

print(arr)
print(copy_arr)
```

```python
# View
view_arr = arr.view()

view_arr[0] = 500

print(arr)
print(view_arr)
```

## 13. Boolean Masking & Filtering

Boolean Masking is a fast way to filter out data in an array by using a list of True and False values

```python
arr = np.array([10, 20, 30, 40, 50])

print(arr > 25)
```

```python
# Filtering
print(arr[arr > 25])
```

```python
numbers = np.array([1, 2, 3, 4, 5, 6, 7, 8])

# Create the mask: Is the remainder 0?
even_mask = (numbers % 2 == 0)

# Filter the array
even_numbers = numbers[even_mask]

print(even_numbers)
# Output: [2, 4, 6, 8]
```

```python
scores = np.array([55, 72, 88, 95, 64, 81, 90])

# Mask: Score must be > 70 AND < 90
b_mask = (scores > 70) & (scores < 90)

b_students = scores[b_mask]

print(b_students)
# Output: [72, 88, 81]
```
