Question 12

Find:

```python
arr = np.array([10, 20, 30, 40, 50])

print("Sum:", arr.sum())
print("Mean:", arr.mean())
print("Max:", arr.max())
print("Min:", arr.min())
```

Question 13

Find all even numbers using boolean masking.

arr = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

```python
arr = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

even=arr % 2 == 0

print(arr[even])
```
