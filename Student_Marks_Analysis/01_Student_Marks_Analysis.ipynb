#1.Basic Analysis



Q1.

Convert the dataset into a NumPy array.

```python
import numpy as np
marks =np.array( [
    [85, 78, 92],
    [70, 88, 80],
    [60, 45, 70],
    [95, 90, 93],
    [40, 50, 65]
])
```

```python
students = np.array([
    "Student 1",
    "Student 2",
    "Student 3",
    "Student 4",
    "Student 5"
])
```

```python
subjects = np.array([
    "Math",
    "Science",
    "English"
])
```

```python
print(marks)
print('\n',students)
print('\n',subjects)
```

##2.Student-wise Analysis

Q4.

Calculate total marks of each student.

```python
print(marks)
total=marks.sum(axis=1)#axis =1 here row wise sum operation
for i in range(total.size):
  print(f'student{i+1} total_marks :{total[i]:.2f}')
```

Q5.

Calculate average marks of each student.

```python
avg_marks=marks.mean(axis=1)
for i in range(avg_marks.size):
  print(f'student{i+1} avg_marks :{avg_marks[i]:.2f}')
```

Q6.

Find the topper of the class based on total marks.

```python
total=marks.sum(axis=1)
print(f'{students[total.argmax()]} is  the topper of class: {total.max()}')
```

Q7.

Find the student with the lowest total marks.

```python
print(f'{students[total.argmin()]} is  the lowest of class: {total.min()}')
```

##Subject-wise Analysis

Q8.

Find highest marks in each subject.

```python
highest_marks=marks.max(axis=0)
for sub , mark in zip(subjects,highest_marks):
  print(f'{sub} : {mark}')
```

Q9.

Find lowest marks in each subject.

```python
lowest_marks=marks.min(axis=0)
for sub , mark in zip(subjects,lowest_marks):
  print(f'{sub} : {mark}')
```

Q10.

Find average marks of each subject.

```python
avg_marks=marks.mean(axis=0)
for sub , mark in zip(subjects,avg_marks):
  print(f'{sub} : {mark}')
```

## Advanced Analysis

Q14.

Find percentage of each student.

```python
sum_mark=marks.sum(axis=1)
print('percentage of each student:')
for stu , sum in zip(students,sum_mark):
  print(f'{stu} :{(sum/300)*100:.2f} %')
```

Q15.

Assign grades using average marks:
| Average Marks | Grade |
| ------------- | ----- |
| 90+           | A     |
| 75-89         | B     |
| 60-74         | C     |
| Below 60      | Fail  |


```python
print("\nGrades:")
avg_mark=marks.mean(axis=1)
for stu , avg in zip(students,avg_mark):
  if avg >= 90:
    print(f'{stu} grade A')
  elif avg >=75:
    print(f'{stu} grade B')
  elif avg>=60:
    print(f'{stu} grade C')
  else:
    print(stu,'grade Fail')
```

Q17.

Find overall class average.

```python
print(f'Overall class average : {marks.mean()}')
```

```python
## using loops
sum=0
count=0
for row in marks:
  for value in row:
    count+=1
    sum+=value
print(f'overall average is {sum/count}')
```

Q18.

Count how many students passed and failed.

```python
# using loops
Passed_Students = 0
Failed_Students = 0
for student in marks:
  if all(student >=50 )==True:
    Passed_Students+=1
  else:
    Failed_Students+=1
print(f'passed students:{Passed_Students} and failed_students:{Failed_Students}')
```

**Q19. Students Sorted by Total Marks**


np.argsort() is a NumPy function that gives the indexes (positions) that would sort an array.

example 1:

arr = np.array([40, 10, 20])

print(np.sort(arr))
#### [10 20 40]

print(np.argsort(arr))
#### [1 2 0]

```python
total_mark=marks.sum(axis=1)
sort_index=total_mark.argsort()[::-1]
rank =1
for ind in sort_index:
  print(f'{rank}. {students[ind]} = {total_mark[ind]}')
  rank+=1
```
